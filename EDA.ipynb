# Exploratory Data Analysis (EDA)

## What is EDA?

**Exploratory Data Analysis (EDA)** is the process of **understanding and analyzing a dataset before applying any Machine Learning model**.

Before building a model, EDA helps us understand:
- what the data looks like
- what patterns exist
- what problems or issues are hidden inside the data

---

## Why Do We Need EDA?

Raw data is often **messy, incomplete, and misleading**.

If we directly apply ML algorithms without EDA:
- the model may learn incorrect patterns
- performance may look good but fail in real-world cases
- hidden issues like outliers or missing values remain unnoticed

EDA helps to:
- understand data distribution
- detect missing and duplicate values
- identify outliers
- find relationships between features
- decide proper preprocessing steps

---

## What Do We Do in EDA?

EDA mainly focuses on answering **four key questions**:

### 1. What is in the dataset?
- number of rows and columns
- feature names
- data types
- numerical vs categorical features

---

### 2. Is the data clean?
- missing values
- duplicate records
- invalid or inconsistent values

---

### 3. How is the data distributed?
Using visualizations like:
- histograms
- box plots
- KDE plots

We analyze:
- skewness
- spread
- outliers

---

### 4. How are features related?
Using:
- scatter plots
- pair plots
- correlation heatmaps

This helps identify:
- strong and weak relationships
- redundant features
- multicollinearity

---

## EDA Is Not Just Visualization

A common misconception is:
> **EDA means only plotting graphs**

In reality, EDA also includes:
- summary statistics
- logical checks
- domain understanding
- forming hypotheses about the data


---

## When Do We Perform EDA?

EDA is performed:
- after loading the dataset
- before data preprocessing
- before model training

## Univariate Analysis

Univariate analysis is the analysis of a single variable to understand its **distribution**, **central tendency**, **spread**, and **data quality**.

---

## Types of Univariate Analysis

### 1. Numerical Variables
Examples: `Age`, `Fare`, `SibSp`, `Parch`

- Measures of central tendency: mean, median, mode  
- Measures of spread: range, variance, standard deviation  
- Distribution shape: skewness, outliers  

Common tools:
- Summary statistics
- Histograms
- Box plots

---

### 2. Categorical Variables
Examples: `Sex`, `Pclass`, `Embarked`, `Survived`

- Frequency counts
- Class imbalance

Common tools:
- Value counts
- Bar charts
- Count plots

---

## What Univariate Analysis Is NOT

- It does **not** explain survival  
- It does **not** show relationships between variables  
- Relationship-based insights come later in **bivariate** and **multivariate** analysis

## Bivariate Analysis

Bivariate analysis studies **two variables at the same time** to understand the **relationship, association, or pattern** between them.  
Most commonly in EDA, one variable is the **target** (e.g. `Survived`) and the other is a **feature**.

---

## What Bivariate Analysis Answers

- Does one variable affect another?
- How does the target change across different feature values?
- Are the variables positively, negatively, or not related?

---

## Types of Bivariate Analysis

### 1. Numerical vs Numerical  
Examples: `Age` vs `Fare`

- Reveals correlation and trends
- Helps detect linear or non-linear relationships

Common tools:
- Scatter plots
- Correlation coefficient (Pearson / Spearman)

---

### 2. Categorical vs Categorical  
Examples: `Sex` vs `Survived`, `Pclass` vs `Survived`

- Shows how categories influence outcomes
- Useful for classification problems

Common tools:
- Grouped bar charts
- Cross-tabulation
- Stacked bar plots

---

### 3. Numerical vs Categorical  
Examples: `Fare` vs `Survived`, `Age` vs `Pclass`

- Compares distributions across categories
- Highlights differences in central tendency and spread

Common tools:
- Box plots
- Violin plots
- Bar plots (mean / median)

---

## What Bivariate Analysis Is NOT

- It does **not** consider interactions among more than two variables  
- It does **not** capture complex multifeature dependencies  
- Those insights come from **multivariate analysis**

---

## One-Line Summary

Bivariate analysis helps understand how **one variable changes with respect to another**, revealing relationships that univariate analysis cannot.

```python
sns.countplot(x='Sex',data=df,hue='Survived')
plt.title("survival count by Gender")
plt.show()
```

```python
sns.countplot(x='Pclass',data=df,hue='Survived')
plt.title("survival by Passenger class")
plt.show()
```

## Multivariate Analysis

Multivariate analysis studies **three or more variables simultaneously** to understand **combined effects, interactions, and complex relationships** within the data.  
It goes beyond pairwise relationships and shows how multiple features work together to influence an outcome.

---

## What Multivariate Analysis Answers

- How do multiple variables jointly affect the target?
- Do variables interact with each other?
- Which features are most important when considered together?
- Are there hidden patterns not visible in univariate or bivariate analysis?

---

## Common Types of Multivariate Analysis

### 1. Multiple Numerical Variables  
Examples: `Age`, `Fare`, `SibSp`

- Reveals correlations among many features
- Helps detect multicollinearity

Common tools:
- Correlation matrix
- Heatmaps
- Pair plots

---

### 2. Mix of Numerical and Categorical Variables  
Examples: `Sex`, `Pclass`, `Age`, `Fare` vs `Survived`

- Shows conditional relationships
- Explains real-world decision patterns

Common tools:
- Grouped box plots
- Facet plots
- Aggregation tables

---

### 3. Model-Based Multivariate Analysis  
Used when visualization is not enough

Common techniques:
- Linear / Logistic Regression
- Decision Trees
- Random Forests
- Feature importance analysis

---

## What Multivariate Analysis Is NOT

- It is **not limited to simple visual inspection**
- It does **not isolate single-variable effects**
- It does **not replace univariate or bivariate analysis**

Multivariate analysis builds **on top of them**, not instead of them.

---

## One-Line Summary

Multivariate analysis examines how **multiple variables interact together** to explain patterns, relationships, and outcomes in complex datasets.

## Feature Engineering

Feature engineering is the process of **creating, transforming, and selecting features** to make machine learning models perform better.  
The goal is simple: **turn raw data into meaningful signals** that models can actually learn from.

---

## Why Feature Engineering Matters

- Raw data is rarely model-ready
- Good features can outperform complex models
- Models donâ€™t understand context â€” features provide it

> Better features often matter more than better algorithms.

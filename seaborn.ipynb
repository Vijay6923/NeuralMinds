## What Is Seaborn?

**Seaborn** is a Python library used for **statistical data visualization**.  
It helps create **attractive, informative, and easy-to-understand plots** with much less code compared to Matplotlib.

Seaborn is built **on top of Matplotlib** and works directly with **Pandas DataFrames**, making it ideal for data analysis and exploratory data analysis (EDA).

It provides high-level functions to visualize:
- Distributions of data  
- Relationships between variables  
- Categorical comparisons  
- Statistical patterns and trends  

Seaborn also comes with **beautiful default themes and color palettes**, which help produce professional-quality plots without extra customization.

## Box Plot

### What is a Box Plot?

A **box plot** (also called a **box-and-whisker plot**) is a graphical summary of a dataset based on **five key statistics**:
- Minimum  
- First Quartile (Q1)  
- Median (Q2)  
- Third Quartile (Q3)  
- Maximum  

---

### Why Do We Use a Box Plot?

Box plots are used because they: 
- Show the **spread and variability** of data  
- Help identify **outliers**    

---

### When Should You Use a Box Plot?

Use a box plot when:
- You want to understand the **distribution** of numerical data  
- You need to detect **outliers**   

#### Examples:
- Exam scores of different classes  
- Salaries across departments  
- Model error comparison across algorithms  


```python
tips.describe()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

## Violin Plot

### What is a Violin Plot?

A **violin plot** is a data visualization technique that combines a **box plot** with a **density plot**.  
It shows not only summary statistics (like median and quartiles) but also the **full distribution shape** of the data.

The width of the violin at any point represents the **density of data values** at that level.

---

### Why Do We Use a Violin Plot?

Violin plots are used because they:
- Show the **entire distribution** of the data  
- Reveal **skewness, multimodality, and spread**  
- Combine the benefits of box plots and density plots

---

### When Should You Use a Violin Plot?

Use a violin plot when:
- You want to understand the **distribution shape** of numerical data  
- You want to compare distributions across multiple categories  
- You suspect skewed or multimodal data  

#### Examples:
- Comparing exam score distributions across classes  
- Salary distributions across departments  
- Model performance distributions  


```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

## KDE Plot (Kernel Density Estimation Plot)

### What is a KDE Plot?

A **KDE plot** (Kernel Density Estimation plot) is a smooth curve that represents the **probability density function** of a continuous numerical variable.  
It is a **smoothed version of a histogram**, where instead of bars, we see a continuous curve showing how data is distributed.

The height of the curve at any point indicates how **dense** the data is around that value.

---

### Why Do We Use a KDE Plot?

KDE plots are used because they:
- Show the **shape of the data distribution** clearly  
- Avoid the bin-size problems of histograms  
- Help identify **skewness, spread, and multiple peaks**  
- Make it easy to compare distributions using multiple KDE curves  

They provide a cleaner and more intuitive view of distribution than raw histograms.

---

### When Should You Use a KDE Plot?

Use a KDE plot when:
- You have **continuous numerical data**  
- You want to understand the **distribution shape**  

#### Examples:
- Exam score distributions  
- Salary distributions  
- Model prediction error distributions  


```python
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
sns.histplot(tips['total_bill'])
```

## Pair Plot

### What is a Pair Plot?

A **pair plot** is a grid of plots that shows **pairwise relationships** between multiple numerical variables in a dataset.  
Each cell in the grid represents a plot between two variables:
- **Scatter plots** for variableâ€“variable relationships  
- **Histograms or KDE plots** on the diagonal for individual variable distributions  


### Why Do We Use a Pair Plot?

Pair plots are used because they:
- Show relationships between **all numerical features at once**  
- Help identify **correlation patterns**  
- Reveal **clusters and trends**  
- Help detect **outliers**  


---

### When Should You Use a Pair Plot?

Use a pair plot when:
- You have **multiple numerical variables**  
- You want a **quick visual overview** of relationships  
- The dataset is **not too large** (otherwise it becomes cluttered)  

#### Examples:
- Exploring features before building an ML model  
- Understanding relationships in datasets like Iris, Wine, or Titanic  
- Comparing variables across different classes using color (`hue`)  

Pair plots are commonly used in **EDA before feature selection and modeling**.

```python
sns.pairplot(tips,hue='sex')
```

## Correlation

### What is Correlation?

**Correlation** is a statistical measure that describes the **strength and direction of the relationship** between two numerical variables.  
It tells us how one variable changes when another variable changes.

The value of correlation lies between **-1 and +1**:
- **+1** â†’ Perfect positive correlation  
- **-1** â†’ Perfect negative correlation  
- **0** â†’ No linear correlation  

---

### Why Do We Use Correlation?

Correlation is used because it helps us:
- Understand **relationships between variables**  
- Identify features that move together  
- Detect **redundant or highly related features**  
- Make better decisions in **feature selection**  

In Machine Learning, correlation helps decide:
- Which features are useful  
- Which features may cause multicollinearity  

---

### When Should You Use Correlation?

Use correlation when:
- Both variables are **numerical**
- You want to measure a **linear relationship**
- You are performing **Exploratory Data Analysis (EDA)**

#### Examples:
- Study hours vs marks  
- Height vs weight  
- Advertising spend vs sales

```python
corr=tips[['total_bill','tip','size']].corr()
```

## Heatmap

### What is a Heatmap?

A **heatmap** is a data visualization technique that represents values in a dataset using **colors**.  
Each cellâ€™s color intensity corresponds to the magnitude of the value it represents.

In Data Science, heatmaps are most commonly used to visualize:
- Correlation matrices  
- Confusion matrices  
- Feature relationships  


---

### Why Do We Use a Heatmap?

Heatmaps are used because they:
- Make large tables of numbers **easy to understand visually**  
- Help quickly identify **high and low values**  
- Reveal **patterns, trends, and relationships**  
- Allow instant comparison using color intensity  

Instead of reading numbers one by one, you can **see patterns at a glance**.

---

### When Should You Use a Heatmap?

Use a heatmap when:
- You want to visualize relationships between many variables  
- You are analyzing a **correlation matrix**  

#### Common Use Cases:
- Correlation between features in a dataset  
- Confusion matrix in classification problems  
- Feature importance or similarity matrices  

Heatmaps are widely used in **Exploratory Data Analysis (EDA)** and **model evaluation** because they provide both clarity and insight.

```python
sns.heatmap(corr,annot=True)
```

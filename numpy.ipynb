Itâ€™s faster than normal Python lists because the numbers are stored in one continuous place in memory, which makes calculations efficient. Most data science and machine learning tools use NumPy at the core for handling numerical data.

```python
## array attributes
arr1=np.array([[1,2,3],[4,5,6]])
print(" array :" , arr1)
print("shape :",arr.shape)
print("size of the array", arr.size)
print("datatype",arr.dtype)
```

```python
## maths function
arr2=np.array([1,2,3,4])
print(np.sqrt(arr2))


print(np.log(arr2))
print(np.sin(arr2))
print(np.exp(arr2))
```

```python
arr3=np.array([1,2,3,4])
mean=np.mean(arr3)
```

```python
mean
```

```python
print(np.median(arr3))
print(np.std(arr3))
print(np.var(arr3))
```

```python
### arithmetic operations
a1=np.array([1,2,3,4,5])
a2=np.array([6,7,8,9,0])
print("addition: ",a1+a2)
print("sub:" ,a1-a2)
print("multiply: ",a1*a2)
print("division :",a1/a2)
```

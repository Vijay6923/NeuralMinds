- **Semantics**:

		•	Refers to the meaning of the code — what it actually does.

		•	A syntactically correct program may still not make semantic sense

		.	It deals with the logic and correct use of types/operations.



	x = "5" + 3  # Syntax is correct, but semantically wrong (TypeError)

```python
print(5/0)
```

```python
a = 10
b = "20"
print(a + b)  # TypeError → semantically incorrect
```

❌ Semantic Error Example:

```python
age = "18"
print(age + 2)  # TypeError: Can't add string and integer
```
